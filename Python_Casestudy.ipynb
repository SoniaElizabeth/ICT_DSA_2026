#Bank Account Manager

```python
class BankAccount:
    def __init__(self, owner, accno, balance):
        self.owner = owner
        self.accno = accno
        self.balance = balance

    def deposit(self, amount):
        if amount> 0:
            self.balance += amount
            print("Amount deposited")
        else:
            print("Invalid amount")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print("Withdrawal successful")

    def check_balance(self):
        print("\n-------- Account Details--------")
        print("Name:", self.owner)
        print("Account Number:", self.accno)
        print("Balance:", self.balance)

#Account Details
name =input("Enter name:")
accno =int(input("Enter account number:"))
balance=float(input("Enter initial balance:"))

account =BankAccount(name, accno, balance)

while True:
    print("1.Deposit")
    print("2.Withdraw")
    print("3.Balance Enquiry")
    print("4.Exit")

    choice =int(input("Enter choice:"))

    if choice== 1:
        amt =float(input("Enter amount:"))
        account.deposit(amt)

    elif choice== 2:
        amt =float(input("Enter amount:"))
        account.withdraw(amt)

    elif choice== 3:
        account.check_balance()

    elif choice== 4:
        print("Thanks")
        break

    else:
        print("Invalid choice")
```

#Weather Data Analyzer

```python
class WeatherAnalyzer:

    def __init__(self, temperatures, humidity, rainfall):
        self.temperatures = temperatures
        self.humidity = humidity
        self.rainfall = rainfall

    # Method to calculate average
    def average(self, data):
        return sum(data) / len(data)

    # Method to determine weather condition
    def weather_condition(self):
        if self.rainfall[-1] > 5:
            return "Rainy"
        elif self.temperatures[-1] > 30:
            return "Sunny"
        else:
            return "Cloudy"

    # Method to display results
    def display(self):
        print("Average Temperature:", self.average(self.temperatures))
        print("Average Humidity:", self.average(self.humidity))
        print("Weather Condition:", self.weather_condition())

# Sample data
temps = [30, 32, 31]
hum = [60, 65, 70]
rain = [0, 10, 2]

# Create object
weather = WeatherAnalyzer(temps, hum, rain)

# Call method
weather.display()
```

```python
class WeatherAnalyzer:
    def __init__(self, region, temperatures, humidity):
        self.region = region
        self.temperatures = temperatures
        self.humidity = humidity

    def calculate_avg_temperature(self):
        return sum(self.temperatures) / len(self.temperatures)

    def calculate_avg_humidity(self):
        return sum(self.humidity) / len(self.humidity)

    def determine_weather(self):
        avg_temp = self.calculate_avg_temperature()
        avg_hum = self.calculate_avg_humidity()

        if avg_temp > 30 and avg_hum < 70:
            return "Sunny"
        elif avg_hum >= 70:
            return "Rainy"
        else:
            return "Cloudy"

    def display_report(self):
        print("\n--- Weather Report ---")
        print("Region:", self.region)
        print("Temperatures:", self.temperatures)
        print("Humidity:", self.humidity)
        print("Average Temperature:", self.calculate_avg_temperature())
        print("Average Humidity:", self.calculate_avg_humidity())
        print("Weather Condition:", self.determine_weather())

1#User Input
region = input("Enter region:")

n = int(input("Enter number of days:"))

temperatures = []
humidity = []

print("\nEnter temperature values:")
for i in range(n):
    temp = float(input(f"Day {i+1}: "))
    temperatures.append(temp)

print("\nEnter humidity values:")
for i in range(n):
    hum = float(input(f"Day {i+1}: "))
    humidity.append(hum)

#Create object
weather = WeatherAnalyzer(region, temperatures, humidity)

#Display result
weather.display_report()
```
